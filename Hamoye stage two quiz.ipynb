```python
# importing relevant libraries that will help
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
df = pd.read_csv('energydata_complete.csv')
df
```

```python
x = df['T2'].values.reshape(-1,1)
y = df['T6'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(x_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred)
print('the r2_score is', round(r2_score, 2))
```

```python
from sklearn.preprocessing import MinMaxScaler

#Firstly, we drop the columns date and lights

df.drop(['date', 'lights'], axis = 1, inplace = True)
```

```python
#Secondly, we normalise our dataset to a common scale using the min max scaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
features_df = normalised_df.drop(columns=[ 'Appliances' ])
target = normalised_df[ 'Appliances' ]
```

```python
#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target,
test_size= 0.3 , random_state= 42 )
```

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) #training the algorithm
```

```python
predicted_value = regressor.predict(x_test)
predicted_value
```

```python
# Question 13 (MAE)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_value)
print('the MAE is ', round(mae, 2))
```

```python
# Question 14 (RSS)
import numpy as np
y_test6=np.array(y_test)
predicted_values6= np.array(predicted_value)
rss = np.sum(np.square(y_test6-predicted_values6))
print('the RSS is ', round(rss,2))
```

```python
# Question 15 (RMSE)
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_value))
print('the RMSE is ', round(rmse, 3))
```

```python
# Question 16 (Coefficient of Determination)
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_value)
print('the Coefficient of Determination is ',round(r2_score, 2))
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha= 0.5 )
ridge_reg.fit(x_train, y_train)
```

```python
# Feature Selection and Lasso Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha= 0.001 )
lasso_reg.fit(x_train, y_train)
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name) :
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = [ 'Features' , col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(regressor, x_train, 'Linear_Model_Weight' )
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight' )
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight' )

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on= 'Features' )
final_weights = pd.merge(final_weights, lasso_weights_df, on= 'Features' )

final_weights
```

```python
print(final_weights.loc[:, 'Linear_Model_Weight'].min())


print(final_weights.loc[:, 'Linear_Model_Weight'].max())
```
